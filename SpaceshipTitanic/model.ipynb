```python
import pandas as pd

trainData = pd.read_csv('train.csv')
testData = pd.read_csv('test.csv')

y = trainData['Transported']
trainData.drop(['Transported','Name'],axis=1,inplace=True)

trainData.head()
```

```python
for col in trainData.columns:
    try:
        print(col,y.corr(trainData[col]))
    except:
        print('Not valid right now for',col)
    
```

```python
from sklearn.model_selection import train_test_split


missing_val_count_by_column = (trainData.isnull().sum())
print(missing_val_count_by_column[missing_val_count_by_column > 0])
```

```python

SetNullToZero = ['RoomService','FoodCourt','FoodCourt','ShoppingMall','Spa','VRDeck']
# Columns with 'Not Applicable'  in NaN
SetNullToNo = ['HomePlanet','Cabin','Destination']
#Columns with median in NaN
SetNullToMed = ['Age']
# Most Common
SetNullToMostFreq = ['CryoSleep','VIP']
```

```python
trainData[SetNullToNo] = trainData[SetNullToNo].fillna(value = 'Not Applicable')
trainData[SetNullToMed] = trainData[SetNullToMed].fillna(trainData[SetNullToMed].median())
trainData[SetNullToZero] = trainData[SetNullToZero].fillna(0,inplace=True)
trainData[SetNullToMostFreq] = trainData[SetNullToMostFreq].fillna(value = trainData[SetNullToMostFreq].mode())

testData[SetNullToNo] = testData[SetNullToNo].fillna(value = 'Not Applicable')
testData[SetNullToMed] = testData[SetNullToMed].fillna(testData[SetNullToMed].median())
testData[SetNullToZero] = testData[SetNullToZero].fillna(0,inplace=True)
testData[SetNullToMostFreq] = testData[SetNullToMostFreq].fillna(value = testData[SetNullToMostFreq].mode())


X_train, X_valid, y_train, y_valid = train_test_split(trainData, y,
                                                    train_size=0.8, test_size=0.2,random_state=0)

X_train = pd.get_dummies(X_train)
X_valid = pd.get_dummies(X_valid)
X_test = pd.get_dummies(testData)
X_train, X_valid = X_train.align(X_valid, join='left', axis=1)
X_train, X_test = X_train.align(X_test, join='left', axis=1)
X_train.head(10)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_error
from xgboost import XGBRegressor


model = XGBRegressor(random_state = 0,n_estimators = 1000,learning_rate = 0.05,n_jobs = 4)

model.fit(X_train,y_train,early_stopping_rounds = 5,eval_set=[(X_valid,y_valid)],verbose = False) # Your code here

# Get predictions
predictions = np.rint(model.predict(X_valid)).astype(bool)

# Calculate MAE
# mae_3 = mean_absolute_error(predictions,y_valid)

# Uncomment to print MAE
# print("Mean Absolute Error:" , mae_3)
# predictions
```

```python
final_predictions = np.rint(model.predict(X_test)).astype(bool)
# 0.4839385624704414
# X_test.head(10)
final_predictions
```

```python
output = pd.DataFrame({'PassengerId': testData.PassengerId,
                        'Transported': final_predictions})
output.to_csv('submission.csv', index=False)
```
